```python
# fit a Linear Regression Model
#build regression model
import statsmodels.formula.api as smf
model1=smf.ols("sunday~daily",data=data1).fit()
```

```python
model1.summary()
```

```python
#Plot the scatter plot and overlay the fitted straight line usibg mtplotlib
x=data1["daily"].values
y=data1["sunday"].values
plt.scatter(x, y, color="m",marker="o",s=30)
b0=13.84
b1=1.33
#Predicted response vector
y_hat=b0+b1*x
#plotting the regestration line
plt.plot(x, y_hat, color="g")
#putting labels
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#Plot the linear regression line using seaborn regplot() method
sns.regplot(x="daily",y="sunday",data=data1)
plt.xlim([0,1250])
plt.show()
```

### Predict for new data points(test data)

```python
#PRedict sunday sales for 200 and 300 daily circulation
newdata=pd.Series([200,300,1500])
```

```python
data_pred=pd.DataFrame(newdata,columns=['daily'])
data_pred
```

```python
model1.predict(data_pred)
```

```python
# Predict on all given training data

pred=model1.predict(data1["daily"])
pred
```

```python
# Addpredicted values as column in data1
data1["Y_hat"]=pred
data1
```

```python
#Compute the error values (Residuals) and add as another column
data1["residuals"]=data1["sunday"]-data1["Y_hat"]
data1
```

```python
# Build regression model
import statsmodels.formula.api as smf
model1=smf.ols("sunday~daily",data=data1).fit()
```

```python
model1.summary()
```

```python
# Compute Mean Squared Error for the model
mse=np.mean((data1["daily"]-data1["Y_hat"])**2)
rmse=np.sqrt(mse)
print("MSE: ",mse)
print("RMSE: ",rmse)
```

```python
#Compute Mean Absolute Error (MAE)
mae=np.mean(np.abs(data1["daily"]-data1["Y_hat"]))
mae
```

### Checking the model rediuls scatter plot (for homoscedaticity) and normality
      -  Check the scatter plot for residuals Vs Y_hat
      -  Check the Q-Q plot for residuals distribution
      -  Check the kde hist plot for residuals
   

```python
#Plot the residuals versus y_hat (to chech wheather residuals are independent of)
plt.scatter(data1["Y_hat"],data1["residuals"])
plt.xlabel("Y_hat")
plt.ylabel("residuals")
```

### Observations:
    - There appears to be no trend and the residuals are randomly placed the zero error line
    - Hence the assumption of homoscedasticity(constant variance in residuals) is satisfied

```python
#Plot the Q-Q plot (to check the normality of residuals)
import statsmodels.api as sm
sm.qqplot(data1["residuals"], line='45', fit=True)
plt.show()
```

```python
# Plot the kde distribution for residuals
sns.histplot(data1["residuals"], kde=True)
```

### Observations:
- The data points are seen to closely follow the reference linne of normality
- Hence the residuals are approximately normally distributed as also can be seen from the kde distribution
- The assumptions of simple linear regression are verified and found to be valid in this case
